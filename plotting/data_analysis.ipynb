```python
fedavg_acc_mnist = {}
fedavg_acc_mnist_dp_01 = {}
fedavg_acc_mnist_dp_05 = {}
fedavg_acc_mnist_dp_ada_01 = {}
fedavg_acc_mnist_dp_ada_05 = {}

for i in range(6):
    fedavg_acc_mnist[i] = res_fedavg_mnist["metrics_centralized"]["test_accuracy"][i][1]
    fedavg_acc_mnist_dp_01[i] = res_fedavg_mnist_dp_01["metrics_centralized"][
        "test_accuracy"
    ][i][1]
    fedavg_acc_mnist_dp_05[i] = res_fedavg_mnist_dp_05["metrics_centralized"][
        "test_accuracy"
    ][i][1]
    fedavg_acc_mnist_dp_ada_01[i] = res_fedavg_mnist_dp_ada_01["metrics_centralized"][
        "test_accuracy"
    ][i][1]
    fedavg_acc_mnist_dp_ada_05[i] = res_fedavg_mnist_dp_ada_05["metrics_centralized"][
        "test_accuracy"
    ][i][1]
```

```python
fedadam_acc_mnist = {}
fedadam_acc_mnist_dp_01 = {}
fedadam_acc_mnist_dp_05 = {}
fedadam_acc_mnist_dp_ada_01 = {}
fedadam_acc_mnist_dp_ada_05 = {}

for i in range(6):
    fedadam_acc_mnist[i] = res_fedadam_mnist["metrics_centralized"]["test_accuracy"][i][
        1
    ]
    fedadam_acc_mnist_dp_01[i] = res_fedadam_mnist_dp_01["metrics_centralized"][
        "test_accuracy"
    ][i][1]
    fedadam_acc_mnist_dp_05[i] = res_fedadam_mnist_dp_05["metrics_centralized"][
        "test_accuracy"
    ][i][1]
    fedadam_acc_mnist_dp_ada_01[i] = res_fedadam_mnist_dp_ada_01["metrics_centralized"][
        "test_accuracy"
    ][i][1]
    fedadam_acc_mnist_dp_ada_05[i] = res_fedadam_mnist_dp_ada_05["metrics_centralized"][
        "test_accuracy"
    ][i][1]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].plot(
    list(fedavg_acc_mnist.keys()), list(fedavg_acc_mnist.values()), label="base"
)
axs[0].plot(
    list(fedavg_acc_mnist_dp_01.keys()),
    list(fedavg_acc_mnist_dp_01.values()),
    label="DP ε=0.1",
    linestyle="dashed",
)
axs[0].plot(
    list(fedavg_acc_mnist_dp_05.keys()),
    list(fedavg_acc_mnist_dp_05.values()),
    label="DP ε=0.5",
    linestyle="dotted",
)
axs[0].plot(
    list(fedavg_acc_mnist_dp_ada_01.keys()),
    list(fedavg_acc_mnist_dp_ada_01.values()),
    label="AdaDPS-U ε=0.1",
    linestyle="dashdot",
)
axs[0].plot(
    list(fedavg_acc_mnist_dp_ada_05.keys()),
    list(fedavg_acc_mnist_dp_ada_05.values()),
    label="AdaDPS-U ε=0.5",
)
axs[0].set_title("FedAvg Test Accuracy Comparison")
axs[0].set_xlabel("Number of Rounds")
axs[0].set_ylabel("Test Accuracy")
axs[0].legend()
axs[0].grid(True)

axs[1].plot(
    list(fedadam_acc_mnist.keys()), list(fedadam_acc_mnist.values()), label="base"
)
axs[1].plot(
    list(fedadam_acc_mnist_dp_01.keys()),
    list(fedadam_acc_mnist_dp_01.values()),
    label="DP ε=0.1",
    linestyle="dashed",
)
axs[1].plot(
    list(fedadam_acc_mnist_dp_05.keys()),
    list(fedadam_acc_mnist_dp_05.values()),
    label="DP ε=0.5",
    linestyle="dotted",
)
axs[1].plot(
    list(fedadam_acc_mnist_dp_ada_01.keys()),
    list(fedadam_acc_mnist_dp_ada_01.values()),
    label="AdaDPS-U ε=0.1",
    linestyle="dashdot",
)
axs[1].plot(
    list(fedadam_acc_mnist_dp_ada_05.keys()),
    list(fedadam_acc_mnist_dp_ada_05.values()),
    label="AdaDPS-U ε=0.5",
)
axs[1].set_title("FedAdam Test Accuracy Comparison")
axs[1].set_xlabel("Number of Rounds")
axs[1].set_ylabel("Test Accuracy")
axs[1].legend()
axs[1].grid(True)

plt.savefig("plots/overallaccuracies.png")

plt.tight_layout()
plt.show()
```
